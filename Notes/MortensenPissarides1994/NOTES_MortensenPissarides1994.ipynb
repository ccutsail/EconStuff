Firm side:
Firms have one job, one of two states, filled and producing or vacant and searching. Job creation means a vacant job has been filled. Destruction happens when job leaves the market. 

Worker side: 
Unemployed and searching or employed and producing. Assume no search happens while producing. Wages are determined by productivity, which is determined by the job. More specifically, the wage is given by $\beta \times$ (surplus from job match), where $\beta$ is just the worker's share of the surplus. Jobs characterized by a 'fixed, irreversable technology.' Price of a unit of whatever is produced is given by $p + \sigma\varepsilon$. All jobs have the same $p$, $\sigma$, while $\varepsilon$ is job-specific.

$$(r+\lambda)S(\varepsilon) = p + \sigma\varepsilon - b + \lambda\int\max[S(x),0]dF(x) -  \beta (vq/u)S(\varepsilon_u) $$

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
font = {'size'   : 22}
matplotlib.rc('font', **font)
np.random.seed(0)
xMin=-10;xMax=100
xDelta=xMax-xMin 
lambda0=10; 
npts = 100
t = np.random.poisson(npts);#Poisson number of points
xx = xDelta*np.random.uniform(0,1,t)+xMin;#x coordinates of Poisson points
upper_envelope = np.zeros_like(xx)
upper_envelope[0] = xx[0]
for i in range(len(xx)):
    if xx[i] > upper_envelope[i-1]:
        upper_envelope[i] = xx[i]
    else:
        upper_envelope[i] = upper_envelope[i-1]

p = 1
sigma = 1
prod_price = p + sigma * xx
plt.figure(figsize=(30,10))
plt.plot(prod_price)
plt.plot(upper_envelope)
plt.plot(np.zeros_like(xx))
plt.xlabel('Time')
plt.ylabel('Product Price (Single Firm)')
plt.title('Single-Firm Prices and Upper Envelope Over Time')
plt.show()
```
